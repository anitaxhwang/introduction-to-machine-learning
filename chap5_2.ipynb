```python
import numpy as np
```

```python
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()
```

```python
y
```

```python
from sklearn.svm import LinearSVR

svm_reg = LinearSVR(epsilon=1.5)
svm_reg.fit(X, y)
```

```python
svm_reg.epsilon
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, y, "bo")
```

```python
x1s = np.linspace(0, 2, 100).reshape(100, 1)
y_pred = svm_reg.predict(x1s)
```

```python
y_pred
```

```python
plt.plot(x1s, y_pred, "k-", linewidth=2, label=r"$\hat{y}$")
plt.plot(x1s, y_pred + svm_reg.epsilon, "k--")
plt.plot(x1s, y_pred - svm_reg.epsilon, "k--")
plt.plot(X, y, "bo")
plt.legend()
```

```python
svm_reg2 = LinearSVR(epsilon=0.5)
svm_reg2.fit(X, y)
y_pred2 = svm_reg2.predict(x1s)
```

```python
plt.plot(x1s, y_pred2, "k-", linewidth=2, label=r"$\hat{y}$")
plt.plot(x1s, y_pred2 + svm_reg2.epsilon, "k--")
plt.plot(x1s, y_pred2 - svm_reg2.epsilon, "k--")
plt.plot(X, y, "bo")
plt.legend()
```

```python
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()
```

```python
from sklearn.svm import SVR

svm_poly_reg1 = SVR(kernel="poly", degree=2, C=100, epsilon=0.1, gamma="auto")
svm_poly_reg2 = SVR(kernel="poly", degree=2, C=0.01, epsilon=0.1, gamma="auto")
svm_poly_reg1.fit(X, y)
svm_poly_reg2.fit(X, y)
```

```python
svm_poly_reg1.epsilon
```

```python
x1s = np.linspace(-1, 1, 100).reshape(100, 1)
y_pred1 = svm_poly_reg1.predict(x1s)
```

```python
plt.plot(x1s, y_pred1, "k-", linewidth=2, label=r"$\hat{y}$")
plt.plot(x1s, y_pred1 + svm_poly_reg1.epsilon, "k--")
plt.plot(x1s, y_pred1 - svm_poly_reg1.epsilon, "k--")
plt.plot(X, y, "bo")
plt.legend()
```

```python
y_pred2 = svm_poly_reg2.predict(x1s)
plt.plot(x1s, y_pred2, "k-", linewidth=2, label=r"$\hat{y}$")
plt.plot(x1s, y_pred2 + svm_poly_reg2.epsilon, "k--")
plt.plot(x1s, y_pred2 - svm_poly_reg2.epsilon, "k--")
plt.plot(X, y, "bo")
plt.legend()
```
